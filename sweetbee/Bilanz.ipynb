```python
import plotly.express as px
import pandas as pd


Stueck = 300
Honigpreis = 0.5
Glaspreis = 0.44 
AmazonBussines = 39 #Pro Monat #Verkaufstarif Einzelanbieter: Keine monatliche Abonnementgebühr
Verkaufsgebüren = .08 #inklusive Artikelpreis und Kosten für Versand und Geschenkverpackung.
# 8 % für Produkte mit einem Gesamtverkaufspreis von bis zu 10,00 €
# 15 % für Produkte mit einem Gesamtverkaufspreis von über 10,00 €
Versandkosten = 2.95 #Standardpaket: 400 g 2,95 € Standardpaket: 900 g 3,38 €
Lagergebüren = 36*0.5 #26,00 €	36,00 € Januar–September	Oktober–Dezember je Kubikmeter und Monat
Verkauf = list(range(1,Stueck))


Honiganzahl = 4

AmazonFix = AmazonBussines + Lagergebüren 
Kosten = (Honigpreis*Honiganzahl + Glaspreis*Honiganzahl + Versandkosten/Honiganzahl)*(1+Verkaufsgebüren)

Preis = [(AmazonFix/x) + Kosten for x in range(1,Stueck)]

MaxPreis=2*Honiganzahl
for i,j in zip(Preis,list(range(len(Preis)))):
    if i<=MaxPreis:
        value = j
        break
    else:
        value = Stueck-2

df = pd.DataFrame(dict(Verkauf=Verkauf,Preis=Preis))
fig = px.line(df, x="Verkauf", y="Preis", title="Honigverkauf pro Monat") 
fig.add_vline(x=value, line_width=3, line_color="green", opacity=0.25)
# fig.add_vrect(x0=150, x1=200, 
#               # annotation_text="sweet start", annotation_position="top left",
#               # annotation=dict(font_size=15, font_family="Times New Roman"),
#               fillcolor="green", opacity=0.25, line_width=0)
fig.add_annotation(x=value, y=25,
            text=f"{Preis[value]:.2f} Euro/{Honiganzahl}St.",
            showarrow=False,
            yshift=20)
fig.add_annotation(x=value, y=25,
            text=f"Verkauft {value*Honiganzahl}St.",
            showarrow=False,
            yshift=5)
fig.update_layout(
    margin=dict(l=20, r=20, t=40, b=20),
     template="plotly_dark",
     plot_bgcolor= "rgba(0, 0, 0, 0)",
     paper_bgcolor= "rgba(0, 0, 0, 0)"
    # paper_bgcolor="LightSteelBlue",
)
# fig.layout.template = 'plotly_dark'
# fig.update_layout({
# 'plot_bgcolor': 'rgba(0, 0, 0, 0)',
# 'paper_bgcolor': 'rgba(255, 255, 255, 0.5)',
# # 'plot_bgcolor': 'rgba(0, 0, 0, 0)',
# # 'paper_bgcolor': 'rgba(0, 0, 0, 0)',
# })
fig.show()
```
